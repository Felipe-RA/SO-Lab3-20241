```python
num_hilos = [1, 2, 4, 8, 16, 32, 64]
iteraciones = list(range(1, 11))  # Número de iteraciones

plt.figure(figsize=(10, 6))

plt.plot(iteraciones, tiempos_un_hilo, marker='o', label='1 hilo')
plt.plot(iteraciones, tiempos_dos_hilos, marker='o', label='2 hilos')
plt.plot(iteraciones, tiempos_cuatro_hilos, marker='o', label='4 hilos')
plt.plot(iteraciones, tiempos_ocho_hilos, marker='o', label='8 hilos')
plt.plot(iteraciones, tiempos_dieciseis_hilos, marker='o', label='16 hilos')
plt.plot(iteraciones, tiempos_treintaidos_hilos, marker='o', label='32 hilos')
plt.plot(iteraciones, tiempos_sesenta_y_cuatro_hilos, marker='o', label='64 hilos')

plt.xlabel('Iteración')
plt.ylabel('Tiempo de ejecución (ms)')
plt.title('Tiempo de ejecución vs Iteración')
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
means = [
    np.mean(tiempos_un_hilo),
    np.mean(tiempos_dos_hilos),
    np.mean(tiempos_cuatro_hilos),
    np.mean(tiempos_ocho_hilos),
    np.mean(tiempos_dieciseis_hilos),
    np.mean(tiempos_treintaidos_hilos),
    np.mean(tiempos_sesenta_y_cuatro_hilos)
]
stds = [
    np.std(tiempos_un_hilo),
    np.std(tiempos_dos_hilos),
    np.std(tiempos_cuatro_hilos),
    np.std(tiempos_ocho_hilos),
    np.std(tiempos_dieciseis_hilos),
    np.std(tiempos_treintaidos_hilos),
    np.std(tiempos_sesenta_y_cuatro_hilos)
]

plt.figure(figsize=(10, 6))
for mean, std, num_hilos in zip(means, stds, [1, 2, 4, 8, 16, 32, 64]):
    x = np.linspace(mean - 3 * std, mean + 3 * std, 100)
    y = np.exp(-0.5 * ((x - mean) / std) ** 2) / (std * np.sqrt(2 * np.pi))
    plt.plot(x, y, label=f'{num_hilos} hilos')

plt.xlabel('Tiempo de ejecución (ms)')
plt.ylabel('Densidad de probabilidad')
plt.title('Distribución Gaussiana de Tiempos de Ejecución')
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
promedio_un_hilo = sum(tiempos_un_hilo) / len(tiempos_un_hilo)
promedio_dos_hilos = sum(tiempos_dos_hilos) / len(tiempos_dos_hilos)
promedio_cuatro_hilos = sum(tiempos_cuatro_hilos) / len(tiempos_cuatro_hilos)
promedio_ocho_hilos = sum(tiempos_ocho_hilos) / len(tiempos_ocho_hilos)
promedio_dieciseis_hilos = sum(tiempos_dieciseis_hilos) / len(tiempos_dieciseis_hilos)
promedio_treintaidos_hilos = sum(tiempos_treintaidos_hilos) / len(tiempos_treintaidos_hilos)
promedio_sesenta_y_cuatro_hilos = sum(tiempos_sesenta_y_cuatro_hilos) / len(tiempos_sesenta_y_cuatro_hilos)

num_hilos = [1, 2, 4, 8, 16, 32, 64]
tiempos_promedio = [
    promedio_un_hilo,
    promedio_dos_hilos,
    promedio_cuatro_hilos,
    promedio_ocho_hilos,
    promedio_dieciseis_hilos,
    promedio_treintaidos_hilos,
    promedio_sesenta_y_cuatro_hilos
]

# Colores para cada serie de datos
colores = ['blue', 'red','green',  'orange','purple', 'brown', 'black']

plt.figure(figsize=(10, 6))
for i in range(len(num_hilos)):
    plt.plot(num_hilos[i], tiempos_promedio[i], marker='o', color=colores[i], label=f'{num_hilos[i]} hilos')
plt.plot(num_hilos, tiempos_promedio, color='pink')


plt.xlabel('Número de hilos')
plt.ylabel('Tiempo de ejecución promedio (ms)')
plt.title('Tiempo de ejecución promedio vs Número de hilos')
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```

Con 16 hilos se lograron tiempos de ejecución menores a los obtenidos con 32 y 64 hilos, siendo estos últimos tiempos de ejecución superiores a los obtenidos con 8 y 4 hilos respectivamente. Esta observación destaca la importancia de encontrar un equilibrio entre la cantidad de hilos utilizados y la eficiencia en el tiempo de ejecución, evitando saturar el sistema con un número excesivo de hilos.

**Distribución Gaussiana de Tiempos de Ejecución:**

Las gráficas de 4, 8, 16 y 32 hilos se mantuvieron muy juntas con bajo tiempo de ejecución y densidad de probabilidad no superior a 0.012. Esto sugiere una consistencia en los tiempos de ejecución y una eficiencia en la gestión de recursos para estos números de hilos.
La gráfica de 32 hilos tuvo un tiempo de ejecución considerablemente mayor y un valor de densidad de probabilidad no superior a 0.006. Esta diferencia destaca una posible ineficiencia en la gestión de recursos con ese número específico de hilos.
Las gráficas de 1 y 2 hilos mostraron tiempos altos y densidades de probabilidad superiores a 0.012, lo que indica que utilizar un número tan bajo de hilos puede resultar en tiempos de ejecución menos eficientes y menos consistentes.

**Tiempo de Ejecución Promedio vs Número de Hilos:**

El tiempo de ejecución más bajo se obtuvo con 16 hilos, siendo el punto más bajo de la parábola. Después de esta cantidad de hilos, comenzó a crecer de nuevo el tiempo de ejecución promedio.
